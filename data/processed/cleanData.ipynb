```python
import pandas as pd

df = pd.read_csv(r"D:\Python Course\Sentiment Analysis on Twitter Project\data\raw\1.6M-Tweets.csv", 
                 encoding='utf-8', encoding_errors='ignore', names=["target", "id", "date", "flag", "user", "text"])
```

Handle Text Col

```python
# Convert to lowercase

df["text"] = df["text"].str.lower()
```

```python
# Remove URL + Mention + Hashtag

df["text"] = (df["text"].str.replace(r'https?://\S+|www\.\S+', '', regex=True)
              .str.replace(r'@\w+', '', regex=True)
              .str.replace(r'#\w+','', regex=True)
              .str.strip())
```

```python
# Remove punctuation & special words + whitespace

df["text"] = (df["text"].str.replace(r'[^A-Za-z0-9\s]','', regex=True)
              .str.replace(r'\s+', ' ', regex=True).str.strip())
```

```python
# Clean extra spaces

df["text"] = df["text"].str.replace(r'\s+', ' ', regex=True).str.strip()
```
